```python
import matplotlib.pyplot as plt
from MeCab import Tagger
t = Tagger()
text = """
任意のテキストを入れる。
"""
splitted = " ".join([x.split("\t")[0] for x in t.parse(text).splitlines()[:-1]])
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
wordcloud = WordCloud(background_color="white"
                      ,font_path="/System/Library/Fonts/ヒラギノ角ゴシック W3.ttc"
                      ,width=800
                      ,height=500).generate(splitted)

plt.imshow(wordcloud)
plt.show()

# wc = WordCloud(background_color="white", font_path="/System/Library/Fonts/ヒラギノ角ゴシック W3.ttc",width=1024,height=674).generate(splitted)
# plt.imshow(wc)
# plt.show()
```
